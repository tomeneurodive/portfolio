```python
import numpy as np
import pandas as pd
from pathlib import Path
from glob import glob

from matplotlib import pyplot as plt
import seaborn as sns
```

グラフの作成

```python

df=AUROC

# Set the style and font sizes
sns.set_style('ticks')
plt.rcParams.update({'font.size': 12})

# Create a Seaborn line plot with different markers for each AUROC
sns.lineplot(x='epoch', y='TrainAUROC_Fold0', data=df, marker='o', color='navy', label='Train_AUROC_Fold0')
sns.lineplot(x='epoch', y='TrainAUROC_Fold1', data=df, marker='s', color='navy', label='Train_AUROC_Fold1')
sns.lineplot(x='epoch', y='TrainAUROC_Fold2', data=df, marker='D', color='navy', label='Train_AUROC_Fold2')
sns.lineplot(x='epoch', y='TrainAUROC_Fold3', data=df, marker='X', color='navy', label='Train_AUROC_Fold3')

sns.lineplot(x='epoch', y='ValidAUROC_Fold0', data=df, marker='o', color='salmon', label='Valid_AUROC_Fold0')
sns.lineplot(x='epoch', y='ValidAUROC_Fold1', data=df, marker='s', color='salmon', label='Valid_AUROC_Fold1')
sns.lineplot(x='epoch', y='ValidAUROC_Fold2', data=df, marker='D', color='salmon', label='Valid_AUROC_Fold2')
sns.lineplot(x='epoch', y='ValidAUROC_Fold3', data=df, marker='X', color='salmon', label='Valid_AUROC_Fold3')


# Set plot title and axes labels
#plt.title('aaa')
plt.xlabel('epochs')
plt.ylabel('AUROC')

# Add a legend
plt.legend(loc='lower right')

# Add a grid
plt.grid(True)

# Remove the top and right spines
#sns.despine()



# Show the plot
plt.show()
```

```python
# Sample data
data = {
    'Months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Product A': [1000, 1500, 1200, 2000, 1800, 2200, 1500, 1800, 2000, 2300, 1900, 2100],
    'Product B': [800, 1200, 1500, 1800, 2000, 2300, 1600, 1900, 2200, 2500, 1800, 2000],
    'Product C': [1200, 1300, 1100, 1400, 1600, 1900, 1300, 1500, 1700, 1900, 1500, 1700]
}
df = pd.DataFrame(data)

# Set the style and font sizes
sns.set_style('ticks')
plt.rcParams.update({'font.size': 12})

# Create a Seaborn line plot with different markers for each product
sns.lineplot(x='Months', y='Product A', data=df, marker='o', color='navy', label='Product A')
sns.lineplot(x='Months', y='Product B', data=df, marker='s', color='teal', label='Product B')
sns.lineplot(x='Months', y='Product C', data=df, marker='D', color='salmon', label='Product C')

# Set plot title and axes labels
plt.title('Sales Across Different Products')
plt.xlabel('Months')
plt.ylabel('Sales ($100K)')

# Add a legend
plt.legend(loc='lower right')

# Add a grid
plt.grid(True)

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```
